```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

df = pd.read_csv('project2.csv')

print(df)

category_counts = df['jobMeta'].value_counts()

plt.figure(figsize=(10, 6))  
category_counts.plot(kind='bar')
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
plt.xlabel('직무')
plt.ylabel('개수')
plt.title('직무별 채용공고 수')
plt.show()
```

```python
category_counts = df[df['career'].str.contains('신입', na=False)]['jobMeta'].value_counts()

plt.figure(figsize=(10, 6)) 
category_counts.plot(kind='bar')
plt.xlabel('직무')
plt.ylabel('개수')
plt.title('직무별 신입 채용공고 수')
plt.show()
```
